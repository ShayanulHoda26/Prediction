```python
def clean(train):
    train = train.drop(['PassengerId','Cabin','Name'],axis=1)

    cols =['RoomService','FoodCourt','ShoppingMall','Spa']
    for col in cols:
        train[col].fillna(train[col].mean(), inplace= True)

    train.Side.fillna('S', inplace= True)
    train.HomePlanet.fillna('Unknown', inplace= True)
    train.Destination.fillna('Unknown', inplace= True)
    train.VIP.fillna('Unknown', inplace= True)
    train.CryoSleep.fillna('False', inplace= True)

    return train

train = clean(train)
```

```python
def clean(test):
    test = test.drop(['PassengerId','Cabin','Name'],axis=1)

    cols =['RoomService','FoodCourt','ShoppingMall','Spa']
    for col in cols:
        test[col].fillna(train[col].mean(), inplace= True)

    test.Side.fillna('S', inplace= True)
    test.HomePlanet.fillna('Unknown', inplace= True)
    test.Destination.fillna('Unknown', inplace= True)
    test.VIP.fillna('Unknown', inplace= True)
    test.CryoSleep.fillna('False', inplace= True)

    return test

test = clean(test)
  
     
```

```python
# Check the lengths of the arrays
len_test_ids = len(test_ids)
len_submission_preds = len(submission_preds)

# Decide how to handle the mismatch
if len_test_ids != len_submission_preds:
    # Truncate or pad arrays to match the length of the shorter one
    min_len = min(len_test_ids, len_submission_preds)
    test_ids = test_ids[:min_len]
    submission_preds = submission_preds[:min_len]
    
# Now both arrays have the same length
df = pd.DataFrame({
    'PassengerId': test_ids.values,
    'Transported': submission_preds,
})
```
